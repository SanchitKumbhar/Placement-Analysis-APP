```python
import pandas as pd

data = pd.read_csv("Placement_Prediction_data.csv")
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.drop("StudentId",axis=1,inplace=True)
```

```python

data.head()
```

```python
data["Internship"]=data["Internship"].replace({"No" : 0,"Yes":1})
data["Hackathon"]=data["Hackathon"].replace({"No" : 0,"Yes":1})
data["PlacementStatus"]=data["PlacementStatus"].replace({"NotPlaced" : 0,"Placed":1})
```

```python
data.info()
```

```python
data.describe()
```

```python
x = data.drop("PlacementStatus",axis=1)
y=data["PlacementStatus"]

y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import  numpy as np
lr=LogisticRegression()
sc = StandardScaler()

sc.fit(x)

x=sc.transform(x)
score=[]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=20)

# pipleline
steps = [("scaler",StandardScaler()),("lr",LogisticRegression())]
pipe = Pipeline(steps)
pipe.fit(xtrain,ytrain)


```

```python
import joblib

joblib.dump(pipe,"model.joblib")
```
