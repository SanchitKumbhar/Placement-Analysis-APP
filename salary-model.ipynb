```python
import pandas as pd

data = pd.read_csv("Salary_prediction_data.csv")
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.drop("StudentId",axis=1,inplace=True)
```

```python
data["Internship"]=data["Internship"].replace({"No" : 0,"Yes":1})
data["Hackathon"]=data["Hackathon"].replace({"No" : 0,"Yes":1})
data["PlacementStatus"]=data["PlacementStatus"].replace({"NotPlaced" : 0,"Placed":1})
```

```python
data=data[data["PlacementStatus"]!=0]

data
```

```python
x = data.drop("salary",axis=1)
y=data["salary"]

y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import *
from sklearn.pipeline import Pipeline
import  numpy as np
lr=Ridge(alpha=0.1)


xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=10)

# pipleline
steps = [("scaler",StandardScaler()),("lr",Ridge(alpha=0.1))]
pipe = Pipeline(steps)
pipe.fit(xtrain,ytrain)
print(ytest)
pipe.predict(xtest)


```

```python
r2 = pipe.score(xtest, ytest)
r2
```

```python
import joblib

joblib.dump(pipe,"model_salary.joblib")
```
