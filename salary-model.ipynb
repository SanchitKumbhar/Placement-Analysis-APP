```python
import pandas as pd

data = pd.read_csv("Salary_prediction_data.csv")
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.drop("StudentId",axis=1,inplace=True)
```

```python
data["Internship"]=data["Internship"].replace({"No" : 0,"Yes":1})
data["Hackathon"]=data["Hackathon"].replace({"No" : 0,"Yes":1})
data["PlacementStatus"]=data["PlacementStatus"].replace({"NotPlaced" : 0,"Placed":1})
```

```python
data=data[data["PlacementStatus"]!=0]

data
```

```python
x = data.drop("salary",axis=1)
y=data["salary"]

y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import *
from sklearn.pipeline import Pipeline
import  numpy as np
lr=LinearRegression()
score =[]

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=274)


# pipleline
steps = [("scaler",StandardScaler()),("lr",LinearRegression())]
pipe = Pipeline(steps)
pipe.fit(xtrain,ytrain)
pipe.predict(xtest)
r2 = pipe.score(xtest, ytest)
score.append(r2)




# Points to be notes:
# 0.8282154542404561 at 274 random state in L2 Ridge algorithm
# 0.828215607979491 at 274 state in L1 Lasso Regression algorithm and Linear Regression
# 0.8282155305347718 at 274 state in Linear Regression

 

 
```
