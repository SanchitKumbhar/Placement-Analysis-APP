```python
import pandas as pd
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Placement_Prediction_data.csv")
```

```python
data.drop("Unnamed: 0",axis=1,inplace=True)
data.drop("StudentId",axis=1,inplace=True)
```

```python

data.head()
```

```python
data["Internship"]=data["Internship"].replace({"No" : 0,"Yes":1})
data["Hackathon"]=data["Hackathon"].replace({"No" : 0,"Yes":1})
data["PlacementStatus"]=data["PlacementStatus"].replace({"NotPlaced" : 0,"Placed":1})
```

```python
data.info()
```

```python
data.describe()
```

```python
x = data.drop("PlacementStatus",axis=1)
y=data["PlacementStatus"]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="CGPA", y="PlacementStatus", hue="Internship", style="Hackathon")
plt.title("Scatterplot of CGPA vs PlacementStatus")
plt.xlabel("CGPA")
plt.ylabel("PlacementStatus")
plt.legend(title="Legend")
plt.show()
```

Logistic regression

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import  numpy as np
lr=LogisticRegression()
sc = StandardScaler()

sc.fit(x)

x=sc.transform(x)
score=[]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=23)


lr.fit(xtrain,ytrain)
print(confusion_matrix(ytest,lr.predict(xtest)))

lr.score(xtest,ytest)
```

DecsionTree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics  import *
dt=DecisionTreeClassifier(max_depth=10)
dt.fit(xtrain,ytrain)
print(dt.score(xtest,ytest))
print(confusion_matrix(ytest,dt.predict(xtest)))

```

Analysis

```python
# prepruning
best=[]
for i in range(1,20):
    dt=DecisionTreeClassifier(max_depth=i)
    dt.fit(xtrain,ytrain)
    best.append(dt.score(xtest,ytest))

print(max(best))
print(np.argmax(best))







```

```python
pipleline
steps = [("scaler",StandardScaler()),("lr",LogisticRegression())]
pipe = Pipeline(steps)
pipe.fit(xtrain,ytrain)
pipe.score(xtest,ytest)


import joblib

joblib.dump(pipe,"model.joblib")
```
